```python
import numpy as np
import pandas as pd
from collections import defaultdict
import pickle
```

```python
data = pd.read_json("train.json")
```

```python
recent_data = data[data['year'] >= 16]
recent_data.head(5)
```

```python
recent_data.shape
```

```python
pro_venue_dict = defaultdict(lambda: defaultdict(int))

for i, row in recent_data.iterrows():
    if pd.isna(row['venue']):
        continue
    for pro_author in row['authors']:
        if pro_author < 100:
            pro_venue_dict[pro_author][row['venue']] += 1
```

```python
pro_venue_dict
```

```python
pro_venue_dict = dict(pro_venue_dict)
```

```python
for key in pro_venue_dict.keys():
    pro_venue_dict[key] = dict(pro_venue_dict[key])
```

```python
pro_venue_dict
```

```python
with open('pro_venue_dict.pickle', 'wb') as handle:
    pickle.dump(pro_venue_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open('pro_venue_dict.pickle', 'rb') as handle:
    b = pickle.load(handle)
```
